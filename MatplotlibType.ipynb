```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 散点图
n = 1024
X = np.random.normal(0, 1, n)
Y = np.random.normal(0, 1, n)
T = np.arctan2(Y, X)        # 设置颜色值XY

plt.scatter(X, Y, s=75, c=T, alpha=0.7)

plt.xlim((-1.5, 1.5))
plt.ylim((-1.5, 1.5))
plt.xticks(())      # 把tick隐藏
plt.yticks(())

plt.show()
```

```python
# 条形图
n = 12
X = np.arange(n)
Y1 = (1-X/float(n))*np.random.uniform(0.5, 1, n)
Y2 = (1-X/float(n))*np.random.uniform(0.5, 1, n)

plt.bar(X, +Y1, facecolor='red', edgecolor='white')
plt.bar(X, -Y2, facecolor='blue', edgecolor='white')

for x, y in zip(X, Y1):
    plt.text(x+0.02, y+0.02, '%.2f' % y, ha='center', va='bottom')
for x, y in zip(X, Y2):
    plt.text(x-0.02, -y-0.02, '%.2f' % -y, ha='center', va='top')
    
plt.xlim(-0.5, n)
plt.ylim(-1.25, 1.25)
plt.xticks(())
plt.yticks(())

plt.show()
```

```python
# 等高线图
def f(x, y):
    return (1-x/2+x**5+y**5)*np.exp(-x**2-y**2)


n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
X, Y = np.meshgrid(x, y)

plt.contourf(X, Y, f(X, Y), 8, alpha=0.75, cmap='hot')     # 等高线的填充颜色
C = plt.contour(X, Y, f(X, Y), 8, colors='black', linewidths=.5)    # 等高线的线
plt.clabel(C, inline=True, fontsize=10)    # 等高线的标签

plt.xticks(())
plt.yticks(())

plt.show()
```

```python
# 画图
a = np.array([0.33, 0.36, 0.31, 0.42, 0.43, 0.52, 0.42, 0.52, 0.65]).reshape((3, 3))

plt.imshow(a, interpolation='nearest', cmap='bone', origin='lower')
plt.colorbar(shrink=0.9)

plt.xticks(())
plt.yticks(())

plt.show()
```

```python
# 3D图形
fig = plt.figure()
ax = Axes3D(fig)

X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.get_cmap('rainbow'))
ax.contourf(X, Y, Z, zdir='z', offset=-2, cmap='rainbow')
ax.set_zlim(-2, 2)

plt.show()
```

```python
# 多个子图方法二
plt.figure()

ax1 = plt.subplot2grid((3, 3), (0, 0), rowspan=1, colspan=3)
ax1.plot([1, 2], [1, 2])
ax1.set_title('ax1_title')
ax2 = plt.subplot2grid((3, 3), (1, 0), rowspan=1, colspan=2)
ax3 = plt.subplot2grid((3, 3), (1, 2), rowspan=2, colspan=1)
ax4 = plt.subplot2grid((3, 3), (2, 0), rowspan=1, colspan=1)
ax5 = plt.subplot2grid((3, 3), (2, 1), rowspan=1, colspan=1)

plt.show()
```

```python
# 图中图
fig = plt.figure()
x = [1, 2, 3, 4, 5, 6, 7]
y = [1, 3, 4, 2, 5, 8, 6]

left, bottom, width, height = [0.1, 0.1, 0.8, 0.8]
ax1 = fig.add_axes([left, bottom, width, height])
ax1.plot(x, y, 'r')
ax1.set_xlabel('x_label')
ax1.set_ylabel('y_label')
ax1.set_title('ax1_subplot')

left, bottom, width, height = [0.2, 0.6, 0.25, 0.25]
ax2 = fig.add_axes([left, bottom, width, height])
ax2.plot(x, y, 'b')
ax2.set_xlabel('x_label')
ax2.set_ylabel('y_label')
ax2.set_title('ax2_subplot')

plt.axes([0.6, .2, .25, .25])
plt.plot(y[::-1], x, 'g')
plt.xlabel('x_label')
plt.ylabel('y_label')
plt.title('ax3_title')

plt.show()
```

```python
# 次坐标轴
x = np.arange(0, 10, 0.1)
y1 = 0.05*x**2
y2 = -1*y1

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(x, y1, 'g-')
ax2.plot(x, y2, 'b--')

ax1.set_xlabel('X data')
ax1.set_ylabel('Y1', color='g')
ax2.set_ylabel('Y2', color='b')

plt.show()
```
