```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 30)
y1 = 2*x + 1
y2 = x**2

plt.figure(num=1)       # figure图像
plt.plot(x, y1)

plt.figure(num=2, figsize=(8, 5))
plt.plot(x, y2)
plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--')

plt.show()
```

```python
x = np.linspace(-3, 3, 30)
y1 = 2*x + 1
y2 = x**2

plt.figure(num=2, figsize=(8, 5))
plt.plot(x, y2)
plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--')

plt.xlim((-2, 3))
plt.ylim((0, 8))
plt.xlabel(' I am X')
plt.ylabel(' I am Y')

new_ticks = np.linspace(-2, 3, 8)
plt.xticks(new_ticks)
plt.yticks([0, 1, 2, 3, 4, 5, 6, 7, 8], list(r'$abcdefghj$'))

plt.show()
```

```python
x = np.linspace(-3, 3, 30)
y1 = 2*x + 1
y2 = x**2

plt.figure(num=2, figsize=(8, 5))       
plt.plot(x, y2)
plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--')

plt.xlim((-2, 3))       # x，y轴设置
plt.ylim((0, 8))
plt.xlabel(' I am X')
plt.ylabel(' I am Y')

new_ticks = np.linspace(-2, 3, 8)        # 刻度设置
plt.xticks(new_ticks)       
plt.yticks([0, 1, 2, 3, 4, 5, 6, 7, 8], list(r'$abcdefghj$'))

# gca = 'get current axis'
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))

plt.show()
```

```python
x = np.linspace(-3, 3, 30)
y1 = 2*x + 1
y2 = x**2

plt.figure(num=2, figsize=(8, 5))
l1, = plt.plot(x, y2, label='up')       # handle要求加，
l2, = plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--', label='down')

plt.xlim((-2, 3))
plt.ylim((0, 8))
plt.xlabel(' I am X')
plt.ylabel(' I am Y')

new_ticks = np.linspace(-2, 3, 8)       
plt.xticks(new_ticks)
plt.yticks([0, 1, 2, 3, 4, 5, 6, 7, 8], list(r'$abcdefghj$'))

plt.legend(handles=[l1, l2], labels=['aaa', 'bbb'], loc='best')     # 图例

plt.show()
```

```python
x = np.linspace(-3, 3, 30)
y = 2*x + 1

plt.figure(num=1, figsize=(8, 5))
plt.plot(x, y)

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))

x0 = 1
y0 = 2*x0 + 1
plt.scatter(x0, y0, s=50, color='b')
plt.plot([x0, x0], [y0, 0], color='k', linestyle='--', linewidth=2.0)

# 标注
plt.annotate(r'$2x+1=%s$' % y0, xy=(x0, y0), xycoords='data', xytext=(+30, -30), textcoords='offset points',
             fontsize=16, arrowprops=dict(arrowstyle='->', connectionstyle='arc3, rad=.2'))

plt.text(-3, 3, r'$this\ is\ the\ some\ text\ \mu$', fontdict={'size': 16, 'color': 'r'})

# tick能见度
for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(12)
    label.set_bbox(dict(facecolor='white', edgecolor='None', alpha=0.7))

plt.show()
```
